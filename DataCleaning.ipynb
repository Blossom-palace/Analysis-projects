#### Import statement

```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/User/fifa21 raw data v2.csv", low_memory=False)
df
```

```python
df.info()
```

#### Remove the playerUrl and photoUrl columns 

```python
df = df.drop(columns=['playerUrl'])
df = df.drop(columns=['photoUrl'])
df
```

#### checking for NULL values

```python
df.isna().sum()
```

```python
df['Hits'].isna().sum()
```

```python
df.duplicated().sum()
```

#### Dropping NULL values

```python
df.dropna(inplace = True)
```

```python
df.info()
```

#### Removing non-digits from Height and Weight columns then converting them to int datatype

```python
df[['Height', 'Weight']] = df[['Height', 'Weight']].replace(r'\D', '', regex=True).astype(int)
df['Height']
```

#### Removing new-line character from the Club column 

```python
df['Club'] = df['Club'].str.replace(r'\n', '', regex=True)
```

```python
df['Club']
```

#### Removing non-digits from 'W/F', 'SM'  and 'IR' columns then converting them to int datatype

```python
df[['W/F', 'SM']] = df[['W/F', 'SM']].replace(r'\D', '', regex=True).astype(int)
df['IR'] = df['IR'].replace(r'\D', '', regex=True).astype(int)
```
